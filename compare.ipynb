we can plot our graphs from a python script from command line. 
On executing the script, it will open a web browser with your Plotly Graph drawn.

```python
import plotly.graph_objs as go
from plotly.offline import init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)
plot([go.Scatter(x=[1, 2, 3], y=[3, 1, 6])])
```

we can also plot your graphs offline inside a Jupyter Notebook Environment. 
First you need to initiate the Plotly Notebook mode as below:

```python
#jupyter setup
init_notebook_mode(connected=True)
```

We can use py.iplot() when working in the Jupyter Notebook to display the plot in the notebook.


```python
iplot([{"x": [1, 2, 3], "y": [3, 1, 6]}])
```

```python
from matplotlib import pyplot as plt
from matplotlib import style
print(style.available)
style.use('grayscale')
x=[5,8,10]
y=[12,16,6]

x2=[6,9,11]
y2=[6,15,7]

plt.plot(x,y,'g',label='line one',linewidth=5)
plt.plot(x2,y2,'c',label='line two',linewidth=5)

plt.legend()
plt.grid(True,color='k')

plt.title('epic info')
plt.ylabel('Y axis')
plt.xlabel('x axis')

plt.show()
```
